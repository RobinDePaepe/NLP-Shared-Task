## Dummy models

```python
# loading data
import os
import pandas as pd

# pipeline
from sklearn.dummy import DummyClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import classification_report
```

```python
#baseline testing

X_dummy = train['text']
y_dummy= train['Language']

dummy_clf_1 = DummyClassifier(strategy="uniform", random_state=1)
dummy_clf_1.fit(X_dummy,y_dummy)
print("Uniform baseline: {}".format(dummy_clf_1.score(X_dummy,y_dummy)))

dummy_clf_2 = DummyClassifier(strategy="most_frequent", random_state=1)
dummy_clf_2.fit(X_dummy,y_dummy)
print("Most frequent baseline: {}".format(dummy_clf_2.score(X_dummy,y_dummy)))

dummy_clf_3 = DummyClassifier(strategy="stratified", random_state=1)
dummy_clf_3.fit(X_dummy,y_dummy)
print("Stratified baseline: {}".format(dummy_clf_3.score(X_dummy,y_dummy)))
```

```python
X_dmodel = train['text']
y_dmodel = train['Language']
# split data in train en test set
X_dmodel_train, X_dmodel_test, y_dmodel_train, y_dmodel_test = train_test_split(X_dmodel, y_dmodel, test_size=0.10, random_state=1)
```

```python
# define vectorizer and classifier
tfvect = CountVectorizer()
tfidfvect = TfidfVectorizer()
svm = SVC()
linsvm = LinearSVC()
```

### Countvectorizer+SVC

```python
# pipeline
pipe = Pipeline([("vec", tfvect), ("cls", svm)])
pipe.fit(X_dmodel_train, y_dmodel_train)
```

```python
# prediction
y_test_pred = pipe.predict(X_dmodel_test)

# accuracy
print(classification_report(y_dmodel_test, y_test_pred))
```

### tfidfvect + SVC

```python
pipe = Pipeline([("vec", tfidfvect), ("cls", svm)])
pipe.fit(X_dmodel_train, y_dmodel_train)
```

```python
# prediction
y_test_pred = pipe.predict(X_dmodel_test)

# accuracy
print(classification_report(y_dmodel_test, y_test_pred))
```

### tfidfvect + linear SVC

```python
pipe = Pipeline([("vec", tfidfvect), ("cls", linsvm)])
pipe.fit(X_dmodel_train, y_dmodel_train)
```

```python
# prediction
y_test_pred = pipe.predict(X_dmodel_test)

# accuracy
print(classification_report(y_dmodel_test, y_test_pred))
```

### cross val tfidfvect + linear SVC

```python
pipe = Pipeline([("vec", tfidfvect), ("cls", linsvm)])
pipe.fit(X_dmodel_train, y_dmodel_train)
```

```python
result_opt_train = cross_validate(pipe, X_dmodel_test, y_dmodel_test, cv=10, scoring=['f1_micro', 'f1_macro'])
result_opt_train
```

#### Proportion of punctuation marks

```python
# create function
def prop_punct(series):
    """Calculates the proportion of the number of punctuation marks compared to the total of POS tags"""
    
    total_pos = len(series.split())
    punct_count = 0
    for tag in series.split():
        if tag == 'PUNCT':
            punct_count += 1
        
    return punct_count/total_pos
```

#### Average sentence length

```python
# create function
def avg_sentlength(series):
    """Calculates the average sentence lengths of a text"""

    sents = re.split('[\.|\?|\!]+', series)
    lengths = []
    
    for s in sents:
        s = s.strip().split()
        length = len(s)
        lengths.append(length)
        
    return mean(lengths)
```

```python
# extract relevant column as series
text_series = nli_tagged['written_text'].astype(str)

# apply function
nli_tagged['avg_sentlength'] = text_series.apply(lambda x: avg_sentlength(x))
```

#### Proportion of capitalized words

```python
# create function
def count_capwords(series):
    """Returns the proportion of capitalized words in a text"""
    
    words = series.split()
    cap_count = 0
    
    for w in words:
        if w[0].isupper():
            cap_count += 1
            
    return cap_count/len(words)
```

```python
# extract relevant column as series
text_series = nli_tagged['written_text'].astype(str)

# apply function
nli_tagged['prop_capwords'] = text_series.apply(lambda x: count_capwords(x))
```

#### Capitalized words at beginning of sentence

```python
# create function
def prop_capwords_beginningsent(series):
    """Calculates the proportion of capitalized words at the beginning of a sentence"""
    
    sents = re.split('[\.|\?|\!]+', series)

    upper_count = 0
    lower_count = 0

    try:
        for s in sents:
            s = s.strip()
        
            if s[0].isupper():
                upper_count += 1
            elif s[0].islower():
                lower_count += 1
            
    except IndexError:
        pass

    total_sents = upper_count + lower_count
    try:
        prop_upper = upper_count / total_sents
        return prop_upper
    except ZeroDivisionError:
        return upper_count / 1
```

```python
# extract relevant column as series
text_series = nli_tagged['written_text'].astype(str)

# apply function
nli_tagged['prop_capwords_beginningsent'] = text_series.apply(lambda x: prop_capwords_beginningsent(x))
```

#### Proportion of capitalized 'I'

```python
# create function
def prop_capI(series):
    """Calculates the proportion of capitalized 'I' vs 'i'"""
    
    words = series.split()
    i_count = 0
    i_upper = 0

    for w in words:
        if w == 'i':
            i_count += 1
        elif w == 'I':
            i_count += 1
            i_upper +=1
    try:
        i_upper_prop = i_upper / i_count
        return i_upper_prop
    except ZeroDivisionError:
        return -1 # was intially None but value cannot be NaN for classifier
```

```python
# extract relevant column as series
text_series = nli_tagged['written_text'].astype(str)

# apply function
nli_tagged['prop_capI'] = text_series.apply(lambda x: prop_capI(x))
```
